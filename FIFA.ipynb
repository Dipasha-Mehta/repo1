```python
from google.colab import files
uploaded=files.upload()

#loading the required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

#loading the fifa matches dataset
fifa=pd.read_csv('fifa.csv')
```

```python
#nationality
fifa['nationality'].value_counts()
```

```python
#top 10 nationality
fifa['nationality'].value_counts()[0:10]
```

```python
#bar plot of top 5 nationality
fifa['nationality'].value_counts()[0:5].keys()
fifa['nationality'].value_counts()[0:5]
plt.figure(figsize=(5,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color="purple")
plt.show()
```

```python
#analysis on column short_name and wage_eur
player_salary=fifa[['short_name','wage_eur']]
player_salary.head()
```

```python
#top 5 players salary
player_salary=player_salary.sort_values(by='wage_eur',ascending=False)
player_salary.head()
```

```python
#bar plot
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'][0:5]),list(player_salary['wage_eur'][0:5]),color=["red","brown","green","blue","orange"])
plt.show()
```

```python
#shooting
#extracting column short_name and shooting

players_shooting=fifa[["short_name","shooting"]]

#sort for the top 5 players
players_shooting.sort_values(by=["shooting"],ascending=False).head()
```

```python
#defending
#extracting column short_name,defending,nationality,club

players_defending=fifa[["short_name","defending","nationality","club_name"]]

#sort for the top 5 players
players_defending.sort_values(by=["defending"],ascending=False).head()
```
